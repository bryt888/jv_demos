### 5. Add Server and DB

### 5.1. **Add a New Server in pgAdmin**
   1. In pgAdmin, click on **Add New Server**.
   2. In the dialog box that opens:
      - **General Tab**:
        - Name: Enter a name for your server (e.g., `Postgres_Server`).
      - **Connection Tab**:
        - Host: Enter `localhost` (or the IP of your server if it's remote).
        - Port: Use `5432` (default for PostgreSQL).
        - Maintenance Database: Enter `postgres`.
        - Username: Enter the PostgreSQL username (e.g., `postgres`).
        - Password: Enter your PostgreSQL password.
      - Click **Save**.

---

### 5.2. **Create a New Database**
   1. Expand your server in the pgAdmin sidebar and right-click on **Databases**.
   2. Click **Create**  **Database**.
   3. In the dialog box:
      - **Database Name**: Enter a name (e.g., `spatial_data`).
      - Owner: Select your username (e.g., `postgres`).
   4. Click **Save**.

### How to Check and Start PostgreSQL on Windows 10:
Yes, you need to ensure that the PostgreSQL database server is running on Windows 10 before using **pgAdmin 4**. If the server is not running, pgAdmin 4 won't be able to connect to your databases.


#### **Method 1: Using Services**
1. Press **Win + R** to open the Run dialog box.
2. Type `services.msc` and press **Enter**.
3. In the Services window:
   - Look for a service named **PostgreSQL-x64-x.x** (where `x.x` is your PostgreSQL version, e.g., `15`).
   - Check the **Status** column:
     - If it says **Running**, PostgreSQL is already running.
     - If it says **Stopped**, you need to start it.
  
4. To start the service:
   - Right-click on the service and select **Start**.
5. Close the Services window and try opening pgAdmin 4 again.

---

#### **Method 2: Using Command Prompt**
1. Open **Command Prompt** as Administrator:
   - Search for `cmd` in the Start menu, right-click it, and choose **Run as administrator**.
2. Start PostgreSQL manually:
   - Run the following command (replace `x.x` with your PostgreSQL version):
     ```cmd
     net start postgresql-x64-x.x
     ```
   - Example:
     ```cmd
     net start postgresql-x64-15
     ```
3. If successful, you'll see a message like:
   ```
   The PostgreSQL-x64-x.x service was started successfully.
   ```

---

#### **Method 3: Automatically Start PostgreSQL**
If you want PostgreSQL to start automatically when you start your computer:
1. Open the **Services** window (`services.msc` as explained earlier).
2. Find the PostgreSQL service, right-click it, and select **Properties**.
3. Set the **Startup Type** to **Automatic** and click **OK**.

---
