### 2. Reproject School locations to wgs84
- we don't want to change the pixels in satellite images from mercator to wgs84
- convert the location of schools from lat/lon to web mercator's meters

### 3. Basic Query Steps
1. look through all pixels in satellite (or radar) images
2. if we find active pixel (like the pixel is heavy rain or any weather events we want), then look at all other locations layer (like school). if the location layer has clients or schools, then send the weather event along with customers in this location (pixel) to message dispatching system to send them out
