### Use PIL lib

```python
from PIL import Image

# Load the image
image = Image.open(r'D:\Tommy\++AI\himawari9\13_6\SAT13_20240828000000_4_13_6.png').convert('RGB')

# Extract the red channel
red_channel = image.split()[0]

# Create a new 8-bit grayscale image
gray_image = Image.new('L', image.size)

# Apply the threshold and keep the original red channel values
for x in range(image.width):
    for y in range(image.height):
        red_value = red_channel.getpixel((x, y))
        if red_value > 170:
            gray_image.putpixel((x, y), red_value)

# Save the result as a PNG file
gray_image.save(r'D:\Tommy\++AI\himawari9\output_image.png')
```

### Use OpenCV

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread(r'D:\Tommy\++AI\himawari9\13_6\SAT13_20240828000000_4_13_6.png')

# Extract the red channel (since it's a gray image, all channels are the same)
red_channel = image[:, :, 2]  # Red channel is at index 2

# Create a mask where the red channel values are greater than 170
mask = red_channel > 170

# Initialize an 8-bit grayscale image with zeros
gray_image = np.zeros_like(red_channel, dtype=np.uint8)

# Apply the mask to keep the original values where the red channel is above 170
gray_image[mask] = red_channel[mask]

# Save the result as a PNG file
cv2.imwrite(r'D:\Tommy\++AI\himawari9\output_gray_image.png', gray_image)
```

```python
import glob
import cv2
import numpy as np
import os
```

```python
files = glob.glob(r'D:\Tommy\++AI\himawari9\13_5\SAT13_*.png')
```

```python
min_val=150

for f in files:
    # Load the image
    image = cv2.imread(f)
    # Extract the red channel (since it's a gray image, all channels are the same)
    red_channel = image[:, :, 2]  # Red channel is at index 2
    # Create a mask where the red channel values are greater than 170
    mask = red_channel > min_val
    # Initialize an 8-bit grayscale image with zeros
    gray_image = np.zeros_like(red_channel, dtype=np.uint8)
    # Apply the mask to keep the original values where the red channel is above 170
    gray_image[mask] = red_channel[mask]
    
    file_name = os.path.basename(f)
    # Save the result as a PNG file
    cv2.imwrite(r'D:\Tommy\++AI\himawari9\13_5_sel'+'\\' + file_name, gray_image)
```
