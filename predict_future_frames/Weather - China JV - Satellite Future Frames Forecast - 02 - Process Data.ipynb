### 2. Solutions for Forecasting Future Frames

- #### Use Band 13 Frames to forecat Band 13 future frames
    - Basically, model is guessing the future frames from past
- #### Use Band 08 and Band 13 Frames to forecat Band 13 future frames
    - Since Band 08 shows upper air vapor, it could help model to increase the accuracy on future frames
- #### Use Band 08, Band 13 Frames and Wind Vector (U,V) from Weather models (like ERA5) to forecat Band 13 future frames
    - Wind direction and speed could help model to predict the wind flow on different elevation based on the model output
    - closest model run will be better.
- #### Could also add Radar data if we can line them up on lat,lon and time    

- #### Since there could be more noise on darker area, remove those area could make the "Strong Cloud" image much cleaner and easy to learn.
