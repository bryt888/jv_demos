### 图片及文件夹：

- **'images_original'**: 包含原始雷达图片
- **'processed'**: 包含人工处理过的雷达图片，杂波全部被清除
- **'mask'**: mask图片，用来表明雷达图片上无杂波的区间

### mask的转换方法： 
#### 1. 将原来的处理过的图片通过open cv将原始图片（32位）转换成灰度图片（8位）
#### 2. 利用numpy 进行转换，将任何灰度照片里面值超过5的设为1， 其它的都设成0

```python
import glob 
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_folder= r'D:\Tommy\++AI\++datasets\jv_radar\images_original'
img_processed = r'D:\Tommy\++AI\++datasets\jv_radar\processed'
img_mask = r'D:\Tommy\++AI\++datasets\jv_radar\mask'
img_check = r'D:\Tommy\++AI\++datasets\jv_radar\check'
```

### 找到所有的图片

```python
files = glob.glob(img_folder+'\\*.png')
```

```python
def generate_check_image(img,mask_img,path_check,opacity):
    # Convert the image from BGR to RGB
    img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    #adjust the transparency of the image
    img2[:,:,0] = img2[:,:,0] * opacity
    img2[:,:,1] = img2[:,:,1] * opacity
    img2[:,:,2] = img2[:,:,2] * opacity
    
    img_masked = img2 * mask_img
    # Blend the images with equal weights
    result = cv2.addWeighted(img2, 0.8, img_masked, 0.8, 0)
    #convert rgb color back to bgr.
    result1 = cv2.cvtColor(result, cv2.COLOR_RGB2BGR)
    #OPENCV will save image in RGB, but assume current data is in BGR, the write function will flip it. this is why we did the flip in above line
    cv2.imwrite(path_check, result1) 
```

```python

for f in files:
    fn = f.split("\\")[-1]
    img = cv2.imread(f)
    print(fn)
    p_name = img_processed+"\\" + fn[0:-4]+"_masked.png"
    #读取处理过图片
    image = cv2.imread(p_name)
    #转换成灰度图片
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    #生成0，1 mask
    mask = np.where(gray > 5, 1, 0)
    #将mask图片保存到mask文件， 文件名要和原始文件一样，方便训练时使用
    m_name = img_mask + "\\" + fn
    cv2.imwrite(m_name, mask) 
    mask_img = cv2.imread(m_name)
    #生成验证图片
    c_name = img_check + "\\" + fn
    generate_check_image(img,mask_img,c_name,0.4)
```
